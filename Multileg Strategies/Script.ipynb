```python
import warnings
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf

warnings.filterwarnings('ignore')
```

# Forex

```python
Forex1 = yf.download('USDBRL=X', start='2000-01-01', auto_adjust=0, back_adjust=0, multi_level_index=0).reset_index()
Forex1['Variat'] = Forex1['Close'].pct_change()
Forex1 = Forex1.dropna(subset=['Variat']) .reset_index(drop=1)
Forex1
```

```python
Forex2 = pd.DataFrame()
Forex2['Date']   = Forex1['Date']
Forex2['Return'] = Forex1['Variat'].mul(100)
Forex2
```

# Index

```python
Index1 = yf.download('^BVSP', start='2000-01-01', auto_adjust=0, back_adjust=0, multi_level_index=0).reset_index()
Index1['Variat'] = Index1['Close'].pct_change()
Index1 = Index1.dropna(subset=['Variat']) .reset_index(drop=1)
Index1
```

```python
Index2 = pd.DataFrame()
Index2['Date']   = Index1['Date']
Index2['Return'] = Index1['Variat'].mul(100)
Index2
```

# Bond

```python
Cetip = pd.read_excel('Storage/Brazil CETIP.xlsx', parse_dates=['Data'])
Cetip['Data'] = Cetip['Data'].apply(pd.to_datetime)
Cetip = Cetip[Cetip['Fator Diário']!='ND'] .reset_index(drop=1)
Cetip
```

```python
Bond2 = pd.DataFrame()
Bond2['Date']    = Cetip['Data']
Bond2['Return']  = Cetip['Fator Diário'].astype(str).apply(lambda x: f'{x[0]}.{x[1:]}').astype(float) *100-100
Bond2
```

# Datas

```python
INI_DATE = max([Forex2['Date'].min(), Index2['Date'].min(), Bond2['Date'].min()])

DATES = np.unique([
    *Forex2[Forex2['Date'] >= INI_DATE]['Date'], 
    *Index2[Index2['Date'] >= INI_DATE]['Date'],
    *Bond2[Bond2['Date'] >= INI_DATE]['Date']
])
```

```python
LONG  = (+1)
SHORT = (-1)
```

```python
Data = pd.DataFrame()
Data['Date']            = DATES
Data['Forex Return']    = Data['Date'].map(Forex2.set_index('Date')['Return'])  .fillna(0)
Data['Index Return']    = Data['Date'].map(Index2.set_index('Date')['Return'])  .fillna(0)
Data['Bond Return']     = Data['Date'].map(Bond2.set_index('Date')['Return'])   .fillna(0)

Data['Combo Return']    = (1*SHORT*Data['Forex Return']) + (1*SHORT*Data['Index Return']) + (2*LONG*Data['Bond Return'])

Data['Forex Acum']      = (1+Data['Forex Return']/100).cumprod().mul(100)
Data['Index Acum']      = (1+Data['Index Return']/100).cumprod().mul(100)
Data['Bond Acum']       = (1+Data['Bond Return'] /100).cumprod().mul(100)
Data['Combo Acum']      = (1+Data['Combo Return']/100).cumprod().mul(100)
Data
```

# Charts

```python
plt.title('-1x BVSP -1x Dollar +2x BR Interest', fontweight='bold')
plt.plot(Data['Date'], Data['Combo Acum'])
plt.show()
```

```python
plt.title('Brazilian Bond Accumulated Return', fontweight='bold')
plt.plot(Data['Date'], Data['Bond Acum'])
plt.show()
```
