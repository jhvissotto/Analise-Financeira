# Load

```python
Raw = pd.read_csv('Storage/Metatrader_WIN.csv', parse_dates=['datetime'])
Raw['datetime'] = pd.to_datetime(Raw['datetime'], errors='coerce')
# Raw.head()
```

```python
def QUERY_SELECT(Raw, SYMBOLS, TFRAMES):
    return Raw[Raw['symbol'].isin(SYMBOLS) & Raw['tf'].isin(TFRAMES)] .reset_index(drop=1)

Query = QUERY_SELECT(Raw, SYMBOLS=['WIN$','WIN$D','WIN$N'], TFRAMES=['M15'])
# Query.head()
```

# Charts - Normal

```python
for SYMBOL, Sec in Query.groupby('symbol', sort=0):

    plt.title(f'{SYMBOL}', fontweight='bold')
    plt.plot(Sec['datetime'], Sec['close']) 
    plt.ylim(bottom=90_000, top=200_000)
    plt.grid(True, alpha=0.33)
    plt.tight_layout()
    plt.savefig(os.path.join(f'Charts (Norm scale)/{SYMBOL}.png'), bbox_inches='tight')
    plt.show()
```

# Charts - Log

```python
for SYMBOL, Sec in Query.groupby('symbol', sort=0):

    plt.title(f'{SYMBOL} (Log scale)', fontweight='bold')
    plt.plot(Sec['datetime'], Sec['close']/Sec['close'].iloc[0])
    plt.yscale('log', base=2)
    plt.ylim(bottom=2**(-0.5), top=2**(+0.5))
    plt.tight_layout()
    plt.savefig(os.path.join(f'Charts (Log scale)/{SYMBOL}.png'), bbox_inches='tight')
    plt.show()
```
